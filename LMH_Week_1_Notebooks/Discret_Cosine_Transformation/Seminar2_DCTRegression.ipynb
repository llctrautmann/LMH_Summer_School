```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#THIS IS A BONUS QUESTION
#DCT regression for images

nb = 28
pi = np.pi
im = np.zeros((28,28))

im[10:15, 10:15]=1
plt.imshow(im)
```

```python
nr,nc = np.shape(im)
xr = np.arange(nr)/nr
xc = np.arange(nc)/nc
X1,X2 = np.meshgrid(xr,xc)
```

```python
plt.imshow(X1)
```

```python
plt.imshow(X2)
```

```python
I = np.arange(nb)
J = np.arange(nb)

X1 = X1[:, :, np.newaxis, np.newaxis]
X2 = X2[:, : ,np.newaxis, np.newaxis]
I = I[np.newaxis, np.newaxis, : , np.newaxis]
J = J[np.newaxis, np.newaxis,  np.newaxis,:]

print('X1 Shape:', X1.shape)
print('X2 Shape:', X2.shape)
print('I Shape:', I.shape)
print('J Shape:', J.shape)
```

```python
basis = np.cos(pi * I * X1) * np.cos(pi * J * X2)
print('Basis Shape [Analytical]:', basis.shape)

basis = np.reshape(basis, (-1, basis.shape[2] * basis.shape[3]))
print('Basis Shape [Reshaped]:', basis.shape)
```

```python
plt.imshow(basis)
```

```python
w = np.linalg.solve(basis.T @ basis, basis.T @ np.reshape(im, (-1, 1)))
print('Coefficients Shape:', w.shape)
```

```python
im_hat = np.reshape(basis @ w, (nr, nc))
plt.imshow(im_hat)
```

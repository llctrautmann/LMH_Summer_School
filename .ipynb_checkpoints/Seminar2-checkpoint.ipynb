```python
#Linear Regression for general functions
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi


x0=np.linspace(-1,1,10)

y0=np.sin(2*pi*.5*x0)+0.3*np.random.randn(x0.size)
plt.plot(x0,y0,"bo")
```

```python
#X= [1 x_1^1, x_1^2 ...x_1^M, 
#.   1, x_2^1, x_2^2, ...x_2^M,
#    1, x_N^1, x_N^2, ..., x_N^M]


#\theta. = (X^TX)^-1 (X^Ty)

nb = 15;
reg = 1e-4
#X = []
#for i in range(nb+1):
#  X.append(x0**i)

x0 = np.reshape(x0,(x0.shape[0],1)) 
powers = np.arange(nb)
powers = powers[np.newaxis,:]
X = x0**powers 
#X = np.array(X).T
print(X.shape)
w = np.linalg.inv(X.T@X + reg*np.eye(nb))@(X.T@y0)
yhat = X@w

error = np.sum((y0-yhat)**2)
print(error/yhat.shape[0])
plt.plot(x0,y0,'bo',x0, yhat)
```
